```python
index_year = {0: '2015', 1: '2016', 2: '2017', 3: '2018'}
```

```python
df2015 = pd.read_excel(path_prefix + "2015pcr.xls")
df2016 = pd.read_excel(path_prefix + "2016pcr.xls")
df2017 = pd.read_excel(path_prefix + "2017pcr.xls")
df2018 = pd.read_excel(path_prefix + "2018pcr.xls")
all_dfs = [df2015, df2016, df2017, df2018]


for index,i in enumerate(all_dfs):
    print("Working on " + index_year[index])
    i["Date of Admission"] = dfbills.loc[dfbills["HOSPITALNO"].isin(i["HOSPITALNO"])]["Date of Admission"]
    i["Date of Discharge"] = dfbills.loc[dfbills["HOSPITALNO"].isin(i["HOSPITALNO"])]["Date of Discharge"]
    i["S.No"] = range(1, len(i) + 1)
    print(i.columns.tolist())
    i.columns = ['Bill Date', 'Hospital Number', 'Name', 'DOB', 'Gender', 'Country', 'Bill No', 'Travel/Trekking Agency', 'Insurance Company', 'Diagnosis', 'Policy Amount', 'USD', 'Billed', 'Total Amount (NPR)', 'Length of Stay', 'optdate', 'Admitted', 'OPD/Admit Date', 'Discharge Date', 'S.No']
    i["Remarks"] = ""
    i["Length of Stay"] = 0
    i["Paid By"] = "Insurance"
    i = i[["S.No", "Hospital Number", "Name", "Bill Date", "OPD/Admit Date", "Discharge Date", "Length of Stay", "Admitted", "Diagnosis", "Travel/Trekking Agency", "Insurance Company", "Total Amount (NPR)", "Paid By", "Remarks"]]
    dfcash = pd.read_excel('pokhara/pkrworking.xlsx', sheet_name=index_year[index])
    total = [i, dfcash]
    total = pd.concat(total)
    print(total[["Hospital Number", "Paid By"]])
    total.columns = ["सि.न", "पर्यटकको नाम", "नाम", "Bill Date" , "भर्ना/OPD मिती", "डिस्चार्ज मिती", "Length of Stay", "Admitted", "उपचारको कारण", "पर्यटक्लाई अस्प्ताल ल्याउने उद्दार कम्पनी", "बीमा कम्पनी", "उपचार रकम", "Paid By", "कैफियत"]
    name_index = total[["पर्यटकको नाम","नाम"]]

    ins = total["Paid By"] == "Insurance"
    total.loc[ins, "बीमा दाबी रकम"] = total[ins]["उपचार रकम"]

    submit = total[["सि.न","पर्यटकको नाम", "भर्ना/OPD मिती", "डिस्चार्ज मिती", "उपचारको कारण", "उपचार रकम","बीमा दाबी रकम", "बीमा कम्पनी", "पर्यटक्लाई अस्प्ताल ल्याउने उद्दार कम्पनी", "कैफियत"]]
    submit = submit.sort_values(by=['पर्यटकको नाम'])
    submit["सि.न"] = range(1, len(submit) + 1)
    submit.to_excel(path_prefix + "submission/" + index_year[index] + 'pkr.xlsx', index=False)
    name_index.to_excel(path_prefix + "submission/" + 'index' + index_year[index] + '.xlsx', index=False)
    
```
